# API Basics

## What is an API? 

* **Definition and Purpose**:
    * An API (Application Programming Interface) is a set of rules and protocols for building and interacting with software applications.
    * APIs allow different software systems to communicate with each other.
    * Analogy: think of a restaurant menu as an API, where the customer (user) can select items (data) without knowing how the kitchen (system) prepares it.


* **APIs in Everyday Life**:
    * Social Media Platforms (e.g., Instagram, X) 
    * Online Shopping (e.g., Amazon, eBay)
    * Streaming Services (e.g., Netflix, YouTube, Spotify)
    
    
## Understanding RESTful APIs

* **REST Principles**:
    * REST (Representational State Transfer) is an architectural style used for designing networked applications.

* How REST uses HTTP requests to access and use data.
    * HTTP Methods: GET (retrieve data), POST (send data), PUT (update data), DELETE (remove data).
        * For instance, if you Google "weather in London," the URL might look like this: https://www.google.com/search?q=weather+in+London.
            * **https://www.google.com/search** is the base URL. 
            * **?q=weather+in+London** is the query string with parameters. 
                * **q** is the parameter name
                * **weather+in+London** is the value you’re asking the server to search for.
        * In this context, the browser is like a client that uses the web server’s API to request and display data.
        * The server’s API processes the GET request and returns the data (in this case, a webpage) that the browser then displays to the user.
    * Status Codes: e.g., 200 OK, 404 Not Found, 500 Internal Server Error.

### **json** library

We can use the **json** library to serialize (convert to a JSON string) and deserialize (convert from a JSON string) our data. 

Data serialized into a JSON string can be easily stored or transmitted (like a file or over a network) and deserialized later. 

```python
import json

# Our Python dictionary (representing Pikachu)
pikachu_dict = {
    "name": "Pikachu",
    "type": "Electric",
    "abilities": ["Static", "Lightning Rod"],
    "stats": {
        "hp": 35,
        "attack": 55,
        "defense": 40,
        "specialAttack": 50,
        "specialDefense": 50,
        "speed": 90
    },
    "evolution": ["Pichu", "Pikachu", "Raichu"]
}

# Convert the Python dictionary to a JSON string
pikachu_json = json.dumps(pikachu_dict, indent=4)
print("JSON String:\n", pikachu_json)
```

```python
pikachu_json['name']
```

```python
# Let's assume pikachu_json is received as a JSON string
print(pikachu_json)

# Convert the JSON string back to a Python dictionary
parsed_data = json.loads(pikachu_json)
print("Parsed Data:\n", parsed_data)
```

```python
parsed_data['name']
```

### Basic API-related terms

* **API (Application Programming Interface)**:
    - PokéAPI (https://pokeapi.co/) provides programmatic access to a massive database of Pokémon-related information. It's like a bridge that lets your code retrieve Pokémon data.
    - Learn to read the API's documentation: https://pokeapi.co/docs/v2
    
* **Endpoint**:
    - An example of an endpoint in PokéAPI is https://pokeapi.co/api/v2/pokemon/. This endpoint allows access to Pokémon data. 
    - If you want to access data about a specific Pokémon, such as Pikachu, the endpoint would be https://pokeapi.co/api/v2/pokemon/pikachu.

* **Parameters**:
    - Parameters are used to specify which data you want. 
    - In PokéAPI, if you want to limit the number of results returned, you might use a URL like https://pokeapi.co/api/v2/pokemon?limit=10 - here, limit=10 is a query parameter.

* **HTTP Methods**:
    - In the case of PokéAPI, you'll primarily use the GET method to retrieve data. 
    - For example, sending a GET request to https://pokeapi.co/api/v2/pokemon/pikachu fetches data about Pikachu.

* **Request**:
    - A request to PokéAPI would be you asking the API for specific Pokémon data. 
    - For instance, requesting the list of abilities of a certain Pokémon. 
    
* **Response**:
    - The data returned by PokéAPI after you make a request. 
    - If you requested Pikachu's data, the response would include all the information the API has on Pikachu, typically in JSON format.
    - PokéAPI returns data in JSON format: {"id": 25, "name": "pikachu", "base_experience": 112, ...}

* **Status Code**:
    - When you make a request to PokéAPI, part of the response will be a status code. If everything goes well, you'll get 200 (OK). If you request data for a Pokémon that doesn’t exist, you might get 404 (Not Found).

That's a lot of information about this Pokémon. We may not need all of this. 

Now, let's create a function to make it easier to get the Pokémon data we really need. 

```python
# Function to fetch and print details about a Pokémon
def get_pokemon_data(pokemon):
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon.lower()}'
    response = requests.get(url)
    
    if response.ok:  # If the request was successful
        data = response.json()
        
        # Print some basic information about the Pokémon
        print(f"Name: {data['name'].title()}")
        print(f"ID: {data['id']}")
        print(f"Height: {data['height']} decimetres")
        print(f"Weight: {data['weight']} hectograms")
        print("Types:")
        for type_entry in data['types']:
            print(f"- {type_entry['type']['name'].title()}")
    else:
        print(f"Failed to retrieve data for {pokemon.title()}.")

# Example usage
pokemon_name = input("Enter the name of the Pokémon: ")
get_pokemon_data(pokemon_name)
```

```python
def get_pokemon_abilities(pokemon):
    # Endpoint for the Pokémon data
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon.lower()}'
    response = requests.get(url)
    
    if response.ok:
        data = response.json()
        
        # Extracting and printing the abilities
        abilities = data['abilities']
        print(f"{pokemon.title()}'s abilities:")
        for ability in abilities:
            # Fetching the ability's details if you want more info 
            # by following the ability's corresponding 'url'
            # This incurs a new API call on the endpoint 'ability'
            ability_response = requests.get(ability['ability']['url'])
            if ability_response.ok:
                ability_data = ability_response.json()
                print(f"- {ability['ability']['name'].title()}: {ability_data['effect_entries'][2]['short_effect']}")
            else:
                print(f"- {ability['ability']['name'].title()}")
    else:
        print(f"Data for {pokemon.title()} not found.")

# Example usage
pokemon_name = input("Enter the name of the Pokémon to retrieve its abilities: ")
get_pokemon_abilities(pokemon_name)
```

## The OpenWeatherMap API

Use Case: Provides weather data, including current weather, forecasts, and historical data.

API Key: Required for access.

Website: https://openweathermap.org/api

```python
import requests

BASE_URL = "http://api.openweathermap.org/data/2.5/weather?"

# Replace 'YOUR_OPENWEATHER_API_KEY' with your actual API key
API_KEY = 'YOUR_OPENWEATHER_API_KEY'

# On Colab, you can save your key(s) and get it when you need it.
# 1. Open Secrets (key icon) in the left sidebar.
# 2. Add a secret named OPENWEATHER_API_KEY and turn on notebook access.
try:
    from google.colab import userdata
    API_KEY = userdata.get('OPENWEATHER_API_KEY')
except ImportError:
    pass  # Not on Colab — use API_KEY from the line above

# Ask for a city name
city = 'Seattle'  # input("Enter a city name: ")

# Update the URL with the city and API key
url = BASE_URL + "q=" + city + "&appid=" + API_KEY + "&units=metric"  # units=metric means Celsius
print(url)

# Send HTTP request
response = requests.get(url)

# Check the status code of the response
if response.ok:
    
    # Retrieve data in JSON format
    data = response.json()
    
    # Get main data block
    main = data['main']
    
    # Extract temperature, pressure, and humidity
    temperature = main['temp']
    pressure = main['pressure']
    humidity = main['humidity']
    
    # Get weather block
    weather = data['weather']
    
    # Weather description
    weather_description = weather[0]['description']
    
    # Print the weather details
    print(f"Weather in {city}:")
    print(f"Temperature: {temperature}°C")
    print(f"Pressure: {pressure} hPa")
    print(f"Humidity: {humidity}%")
    print(f"Description: {weather_description.capitalize()}")
else:
    # Show the error message
    print("Error in the HTTP request")
```
